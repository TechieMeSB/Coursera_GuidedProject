```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_json(r'rain.json')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.plot(x='Month',y='Temperature')
```

```python
df.plot(x='Month',y='Rainfall')
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['Month'],df['Temperature'],label='Temperature')
plt.show()
```

```python
plt.figure(figsize=(17,5))
plt.plot(df['Month'],df['Rainfall'],label='Rainfall')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['Month'],df['Temperature'],label='Temperature')
plt.plot(df['Month'],df['Rainfall'],label='Rainfall')
plt.legend()
plt.show()
```

```python
df1=pd.read_csv('tempYearly.csv')
df1
```

```python
sns.scatterplot(x='Year',y='Temperature',data=df1)
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.regplot(x='Year',y='Temperature',data=df1)
plt.show()
```

```python
df2=pd.read_csv('birthYearly.csv')
df2
```

```python
df2.describe()
```

```python
df2.info()
```

```python
dataP=df2.pivot('month','year','births')
sns.heatmap(dataP,annot=True,fmt='d')
```

```python
sns.jointplot('Rainfall','Temperature',data=df1,kind='reg')
plt.show()
```

```python
sns.jointplot('Rainfall','Temperature',data=df1,kind='hex')
plt.show()
```
